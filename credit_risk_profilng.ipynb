## Count Of Loan Purpose With Default Breakdown:
### Goal: Understand the distribution of loan purposes and how many of those loans defaulted

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='LoanPurpose', hue='Default', palette='Set2')

# Add data labels on each bar segment
for bar in ax.patches:
    height = bar.get_height()
    if height > 0:
        ax.text(
            bar.get_x() + bar.get_width() / 2,    # x position
            height + 0.5,                         # y position
            int(height),                          # label
            ha='center', va='bottom', fontsize=9, fontweight='bold'
        )

plt.title('Loan Purpose Distribution with Default Breakdown', fontsize=14)
plt.xlabel('Loan Purpose', fontsize=12)
plt.ylabel('Number of Loans', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Defaulted', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

## Credit Score Distribution by Loan Default Status
### Goal: See how credit score varies among defaulters vs non-defaulters.

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df[df['Default'] == 0]['CreditScore'], label='Not Defaulted',  fill = True)
sns.kdeplot(data=df[df['Default'] == 1]['CreditScore'], label='Defaulted',  color='red', fill = True)
plt.title('Credit Score Distribution by Loan Default Status')
plt.xlabel('Credit Score')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

## Default Rate by Employment Type
### Goal: Identify which employment types have higher loan default rates.



```python
default_rate_by_emp = df.groupby('EmploymentType')['Default'].mean().reset_index()
default_rate_by_emp['Default'] = default_rate_by_emp['Default'] * 100

plt.figure(figsize=(10, 6))
sns.barplot(data=default_rate_by_emp, x='Default', y='EmploymentType', hue = 'EmploymentType', legend = False,palette='mako')
plt.title('Default Rate by Employment Type')
plt.xlabel('Default Rate (%)')
plt.ylabel('Employment Type')
plt.tight_layout()
plt.show()
```

## Distribution of Credit Score (with bins)
### Goal: Seeing how many applicants fall into risky vs safe credit categories.


```python
bins = [300, 580, 670, 740, 800, 850]
labels = ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent']
df['credit_score_group'] = pd.cut(df['CreditScore'], bins=bins, labels=labels)

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='credit_score_group',hue = 'credit_score_group', legend = False, palette='muted')
plt.title('Credit Score Category Distribution')
plt.xlabel('Credit Score Group')
plt.ylabel('Count')
plt.show()
```

## Default Rate by Employment Type
### Goal: Which employment categories are more likely to default.

```python
default_rate = df.groupby('EmploymentType')['Default'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=default_rate, x='EmploymentType', y='Default',hue = 'EmploymentType', legend= False, palette='coolwarm')
plt.title('Default Rate by Employment Type')
plt.ylabel('Default Rate')
plt.xlabel('Employment Type')
plt.xticks(rotation=45)
plt.show()
```

## Default Rate by Credit Score Range
### Goal: Combine binning and target variable to show risk.

```python
credit_default = df.groupby('credit_score_group')['Default'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=credit_default, x='credit_score_group', y='Default',hue = 'credit_score_group',legend = False, palette='viridis')
plt.title('Default Rate by Credit Score Category')
plt.ylabel('Default Rate')
plt.xlabel('Credit Score Group')
plt.show()
```

## Correlation In Categorical Columns
### We'll be using Label-Encoder encoding for transformation

```python
from sklearn.preprocessing import LabelEncoder

# Make a copy to avoid touching the original
df_encoded = df.copy()

# Select only categorical columns
cat_cols = df_encoded.select_dtypes(include='object').columns

# Encode them
le = LabelEncoder()
for col in cat_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

```python
## Plotting Heatmap:
plt.figure(figsize=(10,8))
sns.heatmap(df_encoded[cat_cols].corr(), annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap of Encoded Categorical Features')
plt.show()
```
